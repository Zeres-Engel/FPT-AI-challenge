```python
from ultralytics import YOLO
import cvzone
import cv2
import math

video_path = 'fires.mp4'  
cap = cv2.VideoCapture(video_path)
model = YOLO('firedetect_model.pt')

# Define new frame dimensions
new_width, new_height = 640, 480

# Get the video FPS
fps = cap.get(cv2.CAP_PROP_FPS)

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('fires.avi', fourcc, fps, (new_width, new_height))

while True:
    ret, frame = cap.read()
    if not ret:
        break  # Kt thc vng lp nu ht video

    frame = cv2.resize(frame, (new_width, new_height))
    results = model(frame, stream=True)

    for info in results:
        boxes = info.boxes
        for box in boxes:
            confidence = math.ceil(box.conf[0] * 100)
            if confidence > 20:
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 5)
                cvzone.putTextRect(frame, f'Fire {confidence}%', [x1 + 8, y1 + 100], scale=1.5, thickness=2)
    
    # Save the frame to the output video
    out.write(frame)

    cv2.imshow('frame', frame)
    if cv2.waitKey(20) & 0xFF == ord('d'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
